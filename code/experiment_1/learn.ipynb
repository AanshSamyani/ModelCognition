```python
import os
```

```python
data_dir_path = "/nlsasfs/home/isea/isea10/aansh/introspection/data/exp1/multi_token_completions"

file_names = os.listdir(data_dir_path)

for i in range(1, 21):
    for file in file_names:
        file_id = int(file.split("_")[-1].split(".")[0])
        if file_id == i:
            print(f"i: {i}, file: {file}")
    
```

```python
per_category_files = {}
for i in range(1, 21):
    for file in os.listdir((data_dir_path)):
        file_id = int(file.split("_")[-1].split(".")[0])
        if file_id == i:
            per_category_files[i] = os.path.join((data_dir_path), file)
```
