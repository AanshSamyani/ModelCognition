```python
import os
import json
import matplotlib.pyplot as plt
```

```python
finetuned_dir = "/nlsasfs/home/isea/isea10/aansh/introspection/results/exp_1_llama_1b/finetuned_multi_token"

base_dir = "/nlsasfs/home/isea/isea10/aansh/introspection/results/exp_1_llama_1b/multi_token"
```

```python
finetuned_accuracy = []
base_accuracy = []
for category in range(1, 21):
    finetuned_metrics_path = os.path.join(finetuned_dir, f"category_{category}", "metrics.json")
    with open(finetuned_metrics_path, "r") as f:
        finetuned_metrics = json.load(f)
    finetuned_accuracy.append(finetuned_metrics["accuracy"])
    
    base_metrics_path = os.path.join(base_dir, f"category_{category}", "metrics.json")
    with open(base_metrics_path, "r") as f:
        base_metrics = json.load(f)
    base_accuracy.append(base_metrics["accuracy"])
```

```python
plt.plot(finetuned_accuracy, label="Finetuned")
plt.plot(base_accuracy, label="Base")
plt.legend()
plt.show()
```

```python
finetuned_accuracy_greater_count = 0
for i in range(len(finetuned_accuracy)):
    if finetuned_accuracy[i] > base_accuracy[i]:
        finetuned_accuracy_greater_count += 1
        print(i+1)
        
print(f"Finetuned accuracy is greater in {finetuned_accuracy_greater_count} out of {len(finetuned_accuracy)} categories.")
```
